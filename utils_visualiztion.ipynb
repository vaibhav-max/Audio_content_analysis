# ploting the histogram of similarity score from th csv file
 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file into a DataFrame
df = pd.read_csv('/data/Root_content/Vaani/audio_content_analysis/audio_matched_files_tsv/similarity_results_unique_pair.csv')

# Plot the histogram of similarity scores
plt.figure(figsize=(10, 6))
sns.histplot(df['Similarity_score'], bins=50, kde=True, color='red')

# Add labels and title
plt.xlabel('Similarity Score')
plt.ylabel('Frequency')
plt.title('Histogram of Similarity Scores')

# Show the plot
plt.show()
```
